```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Age(yrs)'], kde=True)
plt.show()
```

```python
sns.boxplot(x=df['SellPrice'])
plt.show()
```

```python
df['CarModel'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Scatterplot for numeric vs numeric
sns.scatterplot(x='Mileage', y='SellPrice', data=df)
plt.show()
```

While selling price of car increases, the mileage decreases (more price → More powerful engine → more fuel consumption → lower mileage.

```python
# Boxplot for category vs numeric
sns.boxplot(x='CarModel', y='Mileage', data=df)
plt.show()
```

```python
df1= df.drop('CarModel', axis=1)
```

```python
# Correlation heatmap
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
We can say 1) While age of people increases, they choose cars with higher mileage
2) While mileage increases, they choose cars lower higher price
3) While age increases, they tend to buy lower priced cars
```

```python
# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Advanced heatmap (e.g., selective features)
selected_features = ['Age(yrs)', 'Mileage', 'SellPrice']
sns.heatmap(df[selected_features].corr(), annot=True, cmap='viridis')
plt.show()
```
